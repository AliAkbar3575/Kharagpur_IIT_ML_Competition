```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test.csv')
print(df.shape)
df.head(5)
```

```python
dff = df.drop(columns=['RID','departure city', 'departure state','delivery date','segment', 
                       'address code', 'placement date', 'location', 'class'])
dff.dropna(inplace=True)
```

```python
dff['ppi'] = dff['price']/df['no of items']
r = np.max(dff['ppi']) - np.min(dff['ppi'])
dff['ppi'] = (dff['ppi'] - np.mean(dff['ppi']))/r

dff['delivery type'] = dff['delivery type'].map({
    'Standard Class': 4,
    'First Class': 3,
    'Second Class': 2,
    'Same Day': 1
})

dff
```

```python
df1 = pd.get_dummies(dff['discount'])
df2 = pd.get_dummies(dff['sub-class'])
```

```python
catdf = pd.concat([df1, df2], axis=1)
```

```python
l = []
for i in [df1, df2]:
    l.append(i.columns[2])
catdf.drop(columns=l, inplace=True)    
```

```python
test_df = pd.concat([catdf, dff[['delivery type', 'no of items', 'ppi', 'id']]], axis=1)
test_df
```

```python
test_df.to_csv('test data after preprocessing(6).csv', index=False)
```
